```python
x = df[['gmat','gpa','work_experience']].values
y = df['admitted'].values
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
y_train
```

```python
import numpy as np
def sigmoid(x):
  return(1/(1+np.exp(-x)))
x_values_for_sigmoid = np.linspace(-10,10)
plt.figure(figsize=(8,7))
plt.plot(x_values_for_sigmoid,sigmoid(x_values_for_sigmoid),c='teal')
plt.title('Sigmoid funcrion')
plt.grid(True)
plt.text(2,0.3,r'$\sigma(x)=\frac{1}{1+e^{-x}}$',fontsize=26)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
logistic_regression = LogisticRegression()
logistic_regression.fit(x_train,y_train)
y_pred = logistic_regression.predict(x_test)
y_pred
```
